# 5. K-means for compression

## (a)

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# read image from matplotlib

img_large = img.imread('data/mandrill-large.tiff')
plt.figure(figsize=(10, 15))
plt.imshow(img_large)
```

```python
# read image from matplotlib

img_small = img.imread('data/mandrill-small.tiff')
plt.figure(figsize=(10, 15))
plt.imshow(img_small)
```

## (b)

```python
cluster_num = 16
X = img_small.reshape((-1, 3))
X.shape
```

```python
def k_mean(X, cluster_num):
    X = X.astype(np.float32)
    index = np.random.choice(X.shape[0], size=cluster_num, replace=False)
    centroids = X[index]

    err_history = []
    err = 1e6
    while err > 1:
        # calc the min dist for every point
        dist_list = []
        for c in centroids:
            ds = np.sum((X - c) ** 2, axis=1)
            dist_list.append(ds)
        assign = np.vstack(dist_list).argmin(axis=0)

        # calc new centroids
        centroid_list = []
        for i in range(cluster_num):
            centroid_list.append(X[np.where(assign == i)[0]].mean(axis=0))
        centroids_temp = np.vstack(centroid_list)

        err = np.sum(np.abs(centroids_temp - centroids))
        err_history.append(err)
        centroids = centroids_temp
    return centroids, assign, err_history
```

```python
centroids_small, assign_small, err_history_small = k_mean(X, cluster_num)
plt.plot(err_history_small)
plt.xlabel('# iterations')
plt.ylabel('error')
```

## (c) 

```python
X = img_large.reshape((-1, 3))
centroids_large, assign_large, err_history_large = k_mean(X, cluster_num)
```

```python
plt.plot(err_history_large)
plt.xlabel('# iterations')
plt.ylabel('error')
```

```python
compressed = np.zeros(X.shape)
for i in range(cluster_num):
    compressed[np.where(assign_large == i)[0]] = centroids_large[i]
compressed = compressed.reshape(img_large.shape)
compressed = compressed.astype(np.uint8)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 30) )
axes = axes.ravel()

axes[0].imshow(compressed)
axes[0].set_title('Compressed Image')

axes[1].imshow(img_large)
axes[1].set_title('Large Image')

axes[2].imshow(img_small)
axes[2].set_title('Small Image')
```
