```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import math
from scipy.optimize import minimize
from scipy.optimize import fmin_tnc
```

```python
# 1.1 plotting the data
# read the data
score1 = []
score2 = []
temp1 = [[], []]
temp2 = [[], []]
y = []
with open('ex2data1.txt') as f:
    for line in f.readlines():
        score1.append(float(line.split(',')[0]))
        score2.append(float(line.split(',')[1]))
        temp1[int(line.split(',')[2])].append(float(line.split(',')[0]))
        temp2[int(line.split(',')[2])].append(float(line.split(',')[1]))
        y.append(int(line.split(',')[2]))
# plot them
plt.figure(figsize=(12,10))
plt.scatter(temp1[0], temp2[0], c='r', s=80, marker='x', label='not admitted')
plt.scatter(temp1[1], temp2[1], c='g', s=80, marker='o', label='admitted')
plt.legend(loc='upper right', prop={'size': 15})
```

```python
# set data to array
ones = np.ones((len(score1)))
x = np.array((ones, score1, score2)).T
y = np.array(y).reshape((len(y),1))
```

```python
# 1.2 Implementation
# 1.2.1 sigmoid function
def sigmoid(a):
    return 1 / (1 + np.exp(-a))
```

```python
# 1.2.2 cost function and gradient
def getCost(theta, x, y):
    a = sigmoid(np.dot(x, theta))
    cost = (-np.dot(y.T, np.log(a)) - np.dot((1 - y).T, np.log(1 - a))) / x.shape[0]
    return cost
def grad(theta, x, y):
    theta = theta.reshape((x.shape[1], 1))
    a = sigmoid(np.dot(x, theta)) - y
    grad1 = x.T.dot(a) / x.shape[0]
    return grad1.flatten()
```

```python
# 1.2.3 learning parameters using minimize
theta = np.zeros(3)
res = fmin_tnc(func=getCost, x0=theta, args=(x, y), fprime=grad)
theta_e = res[0]

# plot them
x_plt = np.arange(30, 100)
y_plt = (-theta_e[0] - theta_e[1] * x_plt) / theta_e[2]
plt.figure(figsize=(12,10))
plt.scatter(temp1[0], temp2[0], c='r', s=80, marker='x', label='not admitted')
plt.scatter(temp1[1], temp2[1], c='g', s=80, marker='o', label='admitted')
plt.legend(loc='upper right', prop={'size': 15})
plt.plot(x_plt, y_plt)
```

```python
# 1.2.4 evaluate the regression

a = np.array([1, 45, 85])
print(sigmoid(a.dot(theta_e)))
```

```python
# 2. regularized logistic regression
# 2.1 visualize the data
test1 = []
test2 = []
tem1 = [[], []]
tem2 = [[], []]
yd2 = []
with open('ex2data2.txt') as f:
    for line in f.readlines():
        test1.append(float(line.split(',')[0]))
        test2.append(float(line.split(',')[1]))
        tem1[int(line.split(',')[2])].append(float(line.split(',')[0]))
        tem2[int(line.split(',')[2])].append(float(line.split(',')[1]))
        yd2.append(int(line.split(',')[2]))
# plot them
plt.figure(figsize=(12,10))
plt.scatter(tem1[0], tem2[0], c='r', s=80, marker='x', label='not admitted')
plt.scatter(tem1[1], tem2[1], c='g', s=80, marker='o', label='admitted')
plt.legend(loc='upper right', prop={'size': 15})
```

```python
# 2.2 feature mapping
x1 = np.array(test1)
x2 = np.array(test2)
def featMap(x1, x2):
    res = np.ones(x1.shape[0]).reshape((x1.shape[0], 1))
    for i in range(1, 7):
        for j in  range(i + 1):
            a = np.multiply(np.power(x1, i - j), np.power(x2, j)).reshape((x1.shape[0], 1))
            res = np.hstack((res, a))
    return res
xd2 = featMap(x1, x2)
```

```python
# 2.3 cost function
def regGetCost(theta, x, y, lambda_reg):
    return getCost(theta, x, y) + np.sum(np.power(theta, 2)) * lambda_reg / 2 * x.shape[0]
def regGrad(theta, x, y, lambda_reg):
    grad2 = grad(theta, x, y)
    grad2[1:] = grad2[1:] + lambda_reg * theta[1:] / x.shape[0]
    return grad2.flatten()
theta = np.zeros(28)
lambda_reg = 1
yd2 = np.array(yd2).reshape((len(yd2),1))
regGetCost(theta, xd2, yd2, lambda_reg)
```

```python
# 2.3.1 learning parameters using minimize
theta_2 = np.zeros(28)
theta_e2 = fmin_tnc(func=regGetCost, x0=theta_2, args=(xd2, yd2, lambda_reg), fprime=regGrad)[0]
```

```python
# 2.4 Plotting the decision boundary
# lambda = 1
plt.figure(figsize=(12,10))
plt.scatter(tem1[0], tem2[0], c='r', s=80, marker='x', label='not admitted')
plt.scatter(tem1[1], tem2[1], c='g', s=80, marker='o', label='admitted')
plt.legend(loc='upper right', prop={'size': 15})

x_plt2 = np.arange(-0.75, 1.25, 0.05)
y_plt2 = np.arange(-0.75, 1.25, 0.05)
def featMapPlot(x1, x2):
    out = np.ones(1)
    for i in range(1, 7):
        for j in range(i + 1):
            out = np.hstack((out, np.multiply(np.power(x1, i - j), np.power(x2, j))))
    return out
def plot_cont0(x_plt2, y_plt2, theta):
    z_plt2 = np.zeros((len(x_plt2), len(y_plt2)))
    for i in range(len(x_plt2)):
        for j in range(len(y_plt2)):
            z_plt2[i,j] = featMapPlot(x_plt2[i], y_plt2[j]).dot(theta)
    return z_plt2
z_plt2 = plot_cont0(x_plt2, y_plt2, theta_e2)
plt.contour(x_plt2, y_plt2, z_plt2, 0)
```

```python
def plotReg(lambda_plot):
    theta_2 = np.zeros(28)
    theta_e3 = fmin_tnc(func=regGetCost, x0=theta_2, args=(xd2, yd2, lambda_plot), fprime=regGrad)[0]

    # plot it 
    plt.figure(figsize=(12,10))
    plt.scatter(tem1[0], tem2[0], c='r', s=80, marker='x', label='not admitted')
    plt.scatter(tem1[1], tem2[1], c='g', s=80, marker='o', label='admitted')
    plt.legend(loc='upper right', prop={'size': 15})

    x_plt2 = np.arange(-0.75, 1.25, 0.05)
    y_plt2 = np.arange(-0.75, 1.25, 0.05)
    def featMapPlot(x1, x2):
        out = np.ones(1)
        for i in range(1, 7):
            for j in range(i + 1):
                out = np.hstack((out, np.multiply(np.power(x1, i - j), np.power(x2, j))))
        return out
    def plot_cont0(x_plt2, y_plt2, theta):
        z_plt2 = np.zeros((len(x_plt2), len(y_plt2)))
        for i in range(len(x_plt2)):
            for j in range(len(y_plt2)):
                z_plt2[i,j] = featMapPlot(x_plt2[i], y_plt2[j]).dot(theta)
        return z_plt2
    z_plt2 = plot_cont0(x_plt2, y_plt2, theta_e3)
    plt.contour(x_plt2, y_plt2, z_plt2, 0)
```

```python
# lambda = 0
plotReg(0)
```
