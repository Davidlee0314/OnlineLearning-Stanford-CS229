We are gonna prove the K matrix is positive semidefinite.

## (a)
Yes

$u^TKu = u^TK_1u + u^TK_2u$

Since $K_1$ and $K_2$ are kernels, they are positive semidefinite. So $u^TKu \geq 0$

## (b)

No

$u^TKu = u^TK_1u - u^TK_2u$

We don't know whether $u^TKu \geq 0$

## (c)

Yes

Since $a \geq 0,$

$u^TKu = a \cdot (\Sigma_{i=1}^{m}\Sigma_{j=1}^{m}u_i u_j K_{1_{ij}}) \geq 0$

## (d)

No

$u^TKu = -a \cdot (\Sigma_{i=1}^{m}\Sigma_{j=1}^{m}u_i u_j K_{1_{ij}}) \leq 0$

## (e)

Yes

$u^TKu = u^TK_1K_2u = (u^TK_1u)u^T(uu^T)^{-1}K_2u$

Since $A^{-1} = (A^TA)^{-1}A^T,$

replace $A$ with $uu^T$

$(uu^T)^{-1} = ((uu^T)^Tuu^T)^{-1}(uu^T)^T$

and $uu^T$ is symmetric, so 

$(uu^T)^{-1} = ((uu^T)^2)^{-1}uu^T$

$u^TKu = (u^TK_1u)u^T((uu^T)^2)^{-1}u(u^TK_2u)$

Now we have to prove that $((uu^T)^2)^{-1}$ is positive semidefinite.

We know if $A$ is symmetric, then $A = Q^{-1}\Lambda Q^{-1}$

and $A^2 = Q^{-1}\Lambda^2 Q^{-1}$

Since $\Lambda^2$ has all eigenvalues $\geq 0$, $A^2$ is positive semidefinite, and also $A^{2^{-1}}$ 

Replace $A$ with $uu^T$, $(u^TK_1u)(u^T((uu^T)^2)^{-1}u)(u^TK_2u) \geq 0$

## (g)

## (h)

Yes 

First, we need to prove $\forall n,$ if $K$ is a kernel ,then $K^n$ is a kernel too

By induction, $n =0$, $u^TK^0u = u^Tu \geq 0$

Suppose $n=k$ is true, $n=k+1 \rightarrow u^TK^{k+1}u = u^TK^kKu \geq 0$ (from (e))

$u^TKu = \Sigma_{i=1}^{m}\Sigma_{j=1}^{m}u_i u_j \Sigma_{k=1}^{m} a^k K(x^{(i)}, x^{(i)})^k$
