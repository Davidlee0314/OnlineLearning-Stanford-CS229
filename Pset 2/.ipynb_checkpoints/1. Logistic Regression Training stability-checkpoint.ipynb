### (a) 
data A coverges after 30372 iterations, while data B doesn't converge.

```python
# %load lr_debug
from __future__ import division
import numpy as np

try:
    xrange
except NameError:
    xrange = range

def add_intercept(X_):
    m, n = X_.shape
    X = np.zeros((m, n + 1))
    X[:, 0] = 1
    X[:, 1:] = X_
    return X

def load_data(filename):
    D = np.loadtxt(filename)
    Y = D[:, 0]
    X = D[:, 1:]
    return add_intercept(X), Y

def calc_grad(X, Y, theta):
    m, n = X.shape
    grad = np.zeros(theta.shape)

    margins = Y * X.dot(theta)
    probs = 1. / (1 + np.exp(margins))
    grad = -(1./m) * (X.T.dot(probs * Y))

    return grad

def logistic_regression(X, Y):
    m, n = X.shape
    theta = np.zeros(n)
    learning_rate = 10
    theta_list = []
    i = 0
    while True:
        i += 1
        prev_theta = theta
        grad = calc_grad(X, Y, theta)
        theta = theta  - learning_rate * (grad)
        if i % 10000 == 0:
            print('Finished %d iterations' % i)
            theta_list.append(theta)
        if np.linalg.norm(prev_theta - theta) < 1e-15:
            print('Converged in %d iterations' % i)
            break
        if i >= 60000:
            break
    return theta_list

def main():
    print('==== Training model on data set A ====')
    Xa, Ya = load_data('data/data_a.txt')
    logistic_regression(Xa, Ya)

    print('\n==== Training model on data set B ====')
    Xb, Yb = load_data('data/data_b.txt')
    logistic_regression(Xb, Yb)

    return
```

```python
main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dA = np.loadtxt('data/data_a.txt')
dA_x = dA[:,1:3]
dA_y = dA[:,0]
dB = np.loadtxt('data/data_b.txt')
dB_x = dB[:,1:3]
dB_y = dB[:,0]

Xa, Ya = load_data('data/data_a.txt')
Xb, Yb = load_data('data/data_b.txt')
```

```python
# data A
plt.scatter(dA_x[:,0], dA_x[:,1], c = dA_y)
```

```python
theta_list_a = logistic_regression(Xa, Ya)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15,5))
axes = axes.ravel()

for i, t in enumerate(theta_list_a):
    x_plt = np.arange(0, 1.2, 0.2)
    y_plt = (- t[0] - t[1] * x_plt) / t[2]
    axes[i].scatter(dA_x[:,0], dA_x[:,1], c = dA_y)
    axes[i].plot(x_plt, y_plt)
```

```python
# data B
plt.scatter(dB_x[:,0], dB_x[:,1], c = dB_y)
```

```python
theta_list_b = logistic_regression(Xb, Yb)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,10))
axes = axes.ravel()

for i, t in enumerate(theta_list_b):
    x_plt = np.arange(0, 1.2, 0.2)
    y_plt = (- t[0] - t[1] * x_plt) / t[2]
    axes[i].scatter(dB_x[:,0], dB_x[:,1], c = dB_y)
    axes[i].plot(x_plt, y_plt)
    print(i+1, np.sum(np.power(np.array(t), 2)), - t[1] / t[2])
```

From the logistic regression's likelihood function, we want to maximize it

$l(\theta) = \Sigma_{i=1}^m y^{(i)}\log h(x^{(i)}) + (1-y^{(i)})\log (1-h(x^{(i)}))$

whereas $h(x) = \frac{1}{1-e^{-\theta x^{(i)}}}$

However, if the data is separable, we can increase the function by scaling the $||\theta||$. Therefore, only when $||\theta||$ reaches infinity, then the likelihood function will be maximum. From the printed out results, we can see that the $||\theta||$ keep increasing, but the slope isn't changed a lot.
