# 1. Neural Networks: MNIST image classification
## (a)

### The deep learning pipeline
$data_{(B x N)} 
\xrightarrow[(N x H)]{W_1}
z_{1_(B x H)}
\xrightarrow{sigmoid function}
h_{(B x H)}
\xrightarrow[(H x K)]{W_2}
z_{2_{(B x K)}}
\xrightarrow{softmax function}
\hat{y}_{2_{(B x K)}}$



```python
import numpy as np
import matplotlib.pyplot as plt
import copy
%matplotlib inline
```

### Utility functions

```python
def readData(images_file, labels_file):
    x = np.loadtxt(images_file, delimiter=',')
    y = np.loadtxt(labels_file, delimiter=',')
    return x, y

def label_bimatrix(label):
    bimatrix = np.zeros((label.size, 10))
    bimatrix[np.arange(label.size), label.astype(int)] = 1
    return bimatrix

def prepare_data(random_seed=1000):
    trainData, trainLabel = readData('data/images_train.csv', 'data/labels_train.csv')
    trainLabel = label_bimatrix(trainLabel)
    
    # use permutation to make two lists have same orders
    np.random.seed(random_seed)
    p = np.random.permutation(60000)
    trainData = trainData[p,:]
    trainLabel = trainLabel[p,:]
    
    # split data into dev and train sets
    devData = trainData[:10000, :]
    devLabel = trainLabel[:10000, :]
    trainData = trainData[10000:, :]
    trainLabel = trainLabel[10000:, :]
    
    # normalize traindata
    mean = np.mean(trainData)
    std = np.std(trainData)
    trainData = (trainData - mean) / std
    devData = (devData - mean) / std
    
    
    # prepare test data
    testData, testLabel = readData('data/images_test.csv', 'data/labels_test.csv')
    mean = np.mean(testData)
    std = np.std(testData)
    testData = (testData - mean) / std
    testLabel = label_bimatrix(testLabel)
    return trainData, trainLabel, devData, devLabel, testData, testLabel

trainData, trainLabel, devData, devLabel, testData, testLabel = prepare_data()
```

### Calculation functions

-  Calculate backward propagation

$l(W_1, W_2, b_1, b_2) = \frac{1}{m}\Sigma_{i=1}^m\Sigma_{k}^K y_k^{(i)} \log \hat{y}_k^{(i)}$

So for every training sample, we need to calcualte $\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial z_2}$ first

$\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial z_{2_j}} = \Sigma_{k=1}^K \frac{y_k}{\hat{y}_k} \cdot \frac{\partial \hat{y}_k}{\partial z_{2_j}} = 1\{y_k = 1\} \frac{\Sigma_{i}exp(-z_{2_i})}{exp(-z_{2_k})} \cdot \frac{\partial \hat{y}_k}{\partial z_{2_j}}$

$\qquad \qquad = 1\{y_k = 1\} \frac{\Sigma_{i}exp(-z_{2_i})}{exp(-z_{2_k})} \cdot
\frac{ \frac{\partial exp(-z_{2_k})}{\partial z_{2_j}} \cdot  \Sigma_{i}exp(-z_{2_i}) - 
\frac{\partial \Sigma_{i}exp(-z_{2_i})}{\partial z_{2_j}} \cdot exp(-z_{2_k}) } 
{(\Sigma_{i}exp(-z_{2_i}))^2}$

$\qquad \qquad = 1\{y_k = 1\} \frac{1}{exp(-z_{2_k})} \cdot
\frac{1\{k=j\} -exp(-z_{2_k}) \cdot  \Sigma_{i}exp(-z_{2_i}) + 
exp(-z_{2_j}) \cdot exp(-z_{2_k}) } 
{\Sigma_{i}exp(-z_{2_i})}$

$\qquad \qquad = 1\{y_k = 1\} \frac{
exp(-z_{2_j}) - 1\{k = j\}\Sigma_{i}exp(-z_{2_i})}{\Sigma_{i}exp(-z_{2_i})}$

$\qquad \qquad = 1\{y_k = 1\}( \hat{y}_k - 1\{k = j\} )$

Therefore, $\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial z_{2}} = \hat{y} - y$

1. $\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial W_2} = h^T(\hat{y} - y)$

2. $\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial b_2} = \hat{y} - y$

3. $\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial W_1} = x^T(\hat{y} - y)W_2^T \circ h(1 - h)$

4. $\frac{\partial l(W_1, W_2, b_1, b_2)}{\partial b_2} = (\hat{y} - y)W_2^T \circ h(1 - h)$

```python
def sigmoid(x): 
    # masks
    pos_mask = (x >= 0)
    neg_mask = (x < 0)
    
    # calculate
    z = np.zeros_like(x, dtype=float)
    z[pos_mask] = np.exp(-x[pos_mask])
    z[neg_mask] = np.exp(x[neg_mask])
    top = np.ones_like(x, dtype=float)
    top[neg_mask] = z[neg_mask]
    return top / (1 + z)

def softmax(x):
    c = np.max(x, axis=1, keepdims=True)
    numerator = np.exp(x - c)
    denominator = np.sum(numerator, axis=1, keepdims=True)
    return numerator / denominator
    
def update_paras(paras, grad, learning_rate):
    paras['W1'] -= learning_rate * grad['W1']
    paras['W2'] -= learning_rate * grad['W2']
    paras['b1'] -= learning_rate * grad['b1']
    paras['b2'] -= learning_rate * grad['b2']
    
def calc_accuracy(y, labels):
    # predictions 
    pred = np.zeros_like(y)
    pred[np.arange(y.shape[0]), np.argmax(y, axis=1)] = 1
    
    res = np.abs((pred - labels)).sum(axis=1)
    acc = res[res == 0].shape[0] / res.shape[0]
    return acc
```

```python
def backward_prop(data, label, paras):
    h, y_hat, loss = forward_prop(data, label, paras)
    
    # define W, b
    W1 = paras['W1']
    W2 = paras['W2']
    b1 = paras['b1']
    b2 = paras['b2']
    
    # calculate W2, b2 gradients
    delta_2 = (y_hat - label) 
    gradW2 = np.dot(h.T, delta_2)
    gradb2 = np.sum(delta_2, axis=0, keepdims=True)
    
    # calculate W1, b1 gradients
    h_1_h = np.multiply(h, 1 - h)
    delta_1 = np.multiply(np.dot(delta_2, W2.T), h_1_h)
    gradW1 = np.dot(data.T, delta_1)
    gradb1 = np.sum(delta_1, axis=0, keepdims=True)
    
    ## for problem b
    lambda_reg = paras['lambda']
    if lambda_reg > 0:
        gradW2 += lambda_reg * W2
        gradW1 += lambda_reg * W1
    
    grad = {}
    # normalize by batch size!
    B = data.shape[0]
    grad['W1'] = gradW1 / B
    grad['W2'] = gradW2 / B
    grad['b1'] = gradb1 / B
    grad['b2'] = gradb2 / B
    
    return grad
    
def forward_prop(data, label, paras):
    # define W, b
    W1 = paras['W1']
    W2 = paras['W2']
    b1 = paras['b1']
    b2 = paras['b2']
    
    # start propagation
    z1 = data.dot(W1) + b1
    h = sigmoid(z1)
    z2 = h.dot(W2) + b2
    y_hat = softmax(z2)
    
    # calculate the costs
    loss = - np.multiply(label, np.log(y_hat)).sum()
    loss /= data.shape[0]

    return h, y_hat, loss
```

```python
def nn_train(trainData, trainLabel, devData, devLabel, num_hidden=300, \
             learning_rate=5, batch_size=1000, num_epoch=30, lambda_reg=0, paras_reuse={}):
    (M, N) = trainData.shape
    
    # define variables
    H = num_hidden
    K = trainLabel.shape[1]
    B = batch_size
    
    if(lambda_reg == 0):
        paras = {}
        
        # define W, b matrix 
        paras['W1'] = np.random.standard_normal((N, H))
        paras['b1'] = np.zeros((1, H), dtype=float)
        paras['W2'] = np.random.standard_normal((H, K))
        paras['b2'] = np.zeros((1, K), dtype=float)
    else:
        paras = copy.deepcopy(paras_reuse)  
    print(paras)
    
    ## for problem (b)
    paras['lambda'] = lambda_reg
    
    # number of iterations per epoch
    num_iter = int(M / B) 
    tr_loss, tr_acc, dev_loss, dev_acc = [], [], [], []
    
    # start training 
    for i in range(num_epoch):
        print(i, end=',')
        for j in range(num_iter):
            batchData = trainData[j * B: (j + 1)* B, :]
            batchLabel = trainLabel[j * B: (j + 1)* B, :]
            grad = backward_prop(batchData, batchLabel, paras)
            update_paras(paras, grad, learning_rate)
          
        # record the losses
        _h, _y, _loss = forward_prop(trainData, trainLabel, paras)
        tr_loss.append(_loss)
        tr_acc.append(calc_accuracy(_y, trainLabel))
        _h, _y, _loss = forward_prop(devData, devLabel, paras)
        dev_loss.append(_loss)
        dev_acc.append(calc_accuracy(_y, devLabel))

    return paras, tr_loss, tr_acc, dev_loss, dev_acc
```

```python
num_epoch = 30 
paras_noreg, tr_loss, tr_acc, dev_loss, dev_acc = \
nn_train(trainData, trainLabel, devData, devLabel, num_epoch = num_epoch)
```

```python
xs = np.arange(num_epoch)

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax0, ax1, ax2, ax3 = axes.ravel()

loss_diff = np.array(dev_loss)- np.array(tr_loss)
ax0.plot(xs, tr_loss, label='train loss')
ax0.plot(xs, dev_loss, label='dev loss')
ax0.legend()
ax0.set_xlabel('# epoch')
ax0.set_ylabel('CE loss')

ax1.plot(xs, loss_diff)
ax1.set_xlabel('# epoch')
ax1.set_ylabel('Loss Difference')

acc_diff = np.array(tr_acc) - np.array(dev_acc)
ax2.plot(xs, tr_acc, label='train acc')
ax2.plot(xs, dev_acc, label='dev acc')
ax2.legend()
ax2.set_xlabel('# epoch')
ax2.set_ylabel('Accuracy')

ax3.plot(xs, acc_diff)
ax3.set_xlabel('# epoch')
ax3.set_ylabel('Accuracy Difference')
```

## (b)

```python
num_epoch = 30 
lambda_reg = 0.5
paras_reuse = copy.deepcopy(paras_noreg)
paras_reg, tr_loss, tr_acc, dev_loss, dev_acc = \
nn_train(trainData, trainLabel, devData, devLabel, \
         num_epoch = num_epoch, lambda_reg = lambda_reg, paras_reuse=paras_reuse)
```
